```python
import pandas as pd
df=pd.read_excel('weather_data.xls',parse_dates=['day']) 
df.set_index('day',inplace=True) #set index as day columns
df
```

```python
new_df=df.fillna(0)  #replace na with value 0 in the entire dataframe
new_df

```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': 'no event'
})

new_df #change na by different things depending on the columns

```

```python
new_df=df.fillna(method='ffill')
new_df #remplis les trous par la ligne davant, bfill par la ligne dapres
```

```python
new_df=df.fillna(method='ffill',limit=1) #limit of copying missing values
new_df
```

```python
new_df=df.interpolate()
new_df #remplace les nan par une aporximation lineaire, passe de 32 à 28 par 20

new_df=df.interpolate(method='time')
new_df # since the date is closer to the neext cell, the tempearture is closer to the next cell

#estimate missing values
```

```python
new_df=df.dropna(thresh=2)
new_df #destroy row if it doesnt have more than 2 valid values
```
